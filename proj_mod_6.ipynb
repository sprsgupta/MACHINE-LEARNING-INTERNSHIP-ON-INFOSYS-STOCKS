```python
import math
i=1
print('{:<30s}|{:>30s}'.format('Annual Returns','Annual Volatility'))
x=[]
y=[]
z=[]
for ele in li:
    data_curr=pd.DataFrame(ele)
    returns=(data_curr['Close Price'].pct_change())
    daily_returns=returns.mean()*100
    daily_std_dev=returns.std()*100
    #print(daily_returns,daily_std_dev)
  
    annual_mean=daily_returns*252
    annual_stddev=daily_std_dev*(math.sqrt(252))
    z.append(list(set(ele['Symbol']))[0])
    print(i,end=":   ")
    print('{:<30f}|{:>30f}'.format(annual_mean,annual_stddev))
    x.append(annual_mean)
    y.append(annual_stddev)
    
    i+=1
```

```python
#Question 3,4
```

```python
Data=pd.DataFrame()
Data['Returns']=x
Data['Volatility']=y
Data
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

  
#df = DataFrame(Data,columns=['x','y'])
  
kmeans = KMeans(n_clusters=5).fit(Data)
centroids = kmeans.cluster_centers_
#print(centroids)
plt.figure(figsize=(35,20))
plt.scatter(Data.iloc[:,0], Data.iloc[:,1], c= kmeans.labels_.astype(float), s=150, alpha=0.5)
#plt.scatter(Data.iloc[0:10,0], Data.iloc[0:10,1], c= kmeans.labels_.astype(float), s=50, alpha=0.8)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=50,marker='*',alpha=0.3)

for i, txt in enumerate(z):
    plt.annotate(txt, (x[i], y[i]))
plt.grid()
#plt.legend(['Cluster1','Cluster2','Cluster3','Cluster4','Cluster5'])
plt.show()
#print(z)
```

```python
pre=kmeans.predict(Data)
pre=list(pre)

Data['Cluster']=pre
Data
```
